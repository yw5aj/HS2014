```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import string

from createmodel.softtipnaming import getModelNameFromRadiusAndG


class SedDeformed:
    def __init__(self, radius, G, row, col):
        self.radius, self.G = radius, G
        self.row, self.col = row, col
        self.filename = getModelNameFromRadiusAndG(radius, G) + '_sener.png'
        self.filepath = './fig/snapshots/' + self.filename
        self.img = mpimg.imread(self.filepath)
        

radii = np.arange(4e-3, 10e-3, 2e-3)
Gs = np.arange(1e4, 13e4, 4e4)

sedDeformedArray = np.empty((3,3), dtype=object)
for row in range(3):
    for col in range(3):
        sedDeformedArray[row, col] = SedDeformed(radii[row], Gs[col], row, col)
```

```python
fig, axs = plt.subplots(3, 3, figsize=(4.86, 4.86))
#cm = plt.cm.get_cmap('rainbow')

for row in range(3):
    for col in range(3):
        axes = axs[row, col]
        axes.set_aspect(1)
        sedDeformed = sedDeformedArray[row, col]
        imgplot = axes.imshow(sedDeformed.img[:-50, 300:, :], vmin=0, vmax=4.)
        sedDeformed.label = 'G = '+str(int(sedDeformed.G/1e3))+' kPa\nr = '+str(int(sedDeformed.radius*1e3))+' mm'
        axes.set_aspect('equal')
        axes.text(0.22, 1.1, sedDeformed.label, transform=axes.transAxes, va='top', fontsize=8)
        axes.axis('off')
        axes.text(.1, 1.0, string.ascii_uppercase[row*3+col], transform=axes.transAxes, fontsize=12, fontweight='bold')

cax = fig.add_axes([0.3, 0.1, 0.4, 0.02])
cbar = fig.colorbar(imgplot, cax=cax, orientation='horizontal')
cbar.set_label(r'SED (kJ/$m^3$)')

fig.subplots_adjust(left=0, right=1, bottom=0.1, top=1, wspace=0, hspace=0)
fig.savefig('./fig/sed_deformed.png', format='png', dpi=300)
```
